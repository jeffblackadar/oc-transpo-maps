```r
library(R6)


RLeafletMap <- R6Class("RLeafletMap",
                  public = list(
                    outputFileHtmlCon = NULL,
                    markers = 0,

                    initialize = function(outputFileHtmlCon) {
                      self$outputFileHtmlCon <- outputFileHtmlCon
                      self$markers <- 0
                    },
                    genMapHTMLTop = function() {
                      writeLines(paste0('<!DOCTYPE html>'), self$outputFileHtmlCon)
                      writeLines(paste0('<html>'), self$outputFileHtmlCon)
                      writeLines(paste0('<head><meta http-equiv="Content-Type" content="text/html; charset=utf-8">'), self$outputFileHtmlCon)
                      writeLines(paste0('<title>map title</title>'), self$outputFileHtmlCon)
                      writeLines(paste0('<link href="http://cdn.leafletjs.com/leaflet-0.6.4/leaflet.css" rel="stylesheet" /><script src="http://cdn.leafletjs.com/leaflet-0.6.4/leaflet.js"></script><script src="https://code.jquery.com/jquery-2.1.4.min.js"></script>'), self$outputFileHtmlCon)
                      writeLines(paste0('<style type="text/css">#my-map {'), self$outputFileHtmlCon)
                      writeLines(paste0('  width:960px;'), self$outputFileHtmlCon)
                      writeLines(paste0('  height:700px;'), self$outputFileHtmlCon)
                      writeLines(paste0('}'), self$outputFileHtmlCon)
                      writeLines(paste0('h1 {'), self$outputFileHtmlCon)
                      writeLines(paste0('  color: #AA0000;'), self$outputFileHtmlCon)
                      writeLines(paste0('  font-size: 24px;'), self$outputFileHtmlCon)
                      writeLines(paste0('  font-family: verdana;'), self$outputFileHtmlCon)
                      writeLines(paste0('}'), self$outputFileHtmlCon)
                      writeLines(paste0('h2 {'), self$outputFileHtmlCon)
                      writeLines(paste0('  color: #000000;'), self$outputFileHtmlCon)
                      writeLines(paste0('  font-size: 20px;'), self$outputFileHtmlCon)
                      writeLines(paste0('  font-family: verdana;'), self$outputFileHtmlCon)
                      writeLines(paste0('}'), self$outputFileHtmlCon)
                      writeLines(paste0('h3 {'), self$outputFileHtmlCon)
                      writeLines(paste0('  color: #000000;'), self$outputFileHtmlCon)
                      writeLines(paste0('  font-size: 16px;'), self$outputFileHtmlCon)
                      writeLines(paste0('  font-family: verdana;'), self$outputFileHtmlCon)
                      writeLines(paste0('}'), self$outputFileHtmlCon)    
                      writeLines(paste0('p {'), self$outputFileHtmlCon)
                      writeLines(paste0('  color: #996600;'), self$outputFileHtmlCon)
                      writeLines(paste0('  font-family: verdana;'), self$outputFileHtmlCon)
                      writeLines(paste0('}'), self$outputFileHtmlCon)
                      writeLines(paste0('li {'), self$outputFileHtmlCon)
                      writeLines(paste0('  color: #996600;'), self$outputFileHtmlCon)
                      writeLines(paste0('  font-family: verdana;'), self$outputFileHtmlCon)
                      writeLines(paste0('}'), self$outputFileHtmlCon)  
                      writeLines(paste0('th {'), self$outputFileHtmlCon)
                      writeLines(paste0('  text-align: right;'), self$outputFileHtmlCon)
                      writeLines(paste0('  font-family: verdana;'), self$outputFileHtmlCon)
                      writeLines(paste0('  color: #663300;'), self$outputFileHtmlCon)
                      writeLines(paste0('}'), self$outputFileHtmlCon)
                      writeLines(paste0('td {'), self$outputFileHtmlCon)
                      writeLines(paste0('  text-align: right;'), self$outputFileHtmlCon)
                      writeLines(paste0('  font-family: verdana;'), self$outputFileHtmlCon)
                      writeLines(paste0('  color: #663300;'), self$outputFileHtmlCon)
                      writeLines(paste0('}'), self$outputFileHtmlCon)
                      writeLines(paste0('</style>'), self$outputFileHtmlCon)
                      writeLines(paste0('</head>'), self$outputFileHtmlCon)
                      writeLines(paste0('<body>'), self$outputFileHtmlCon)
                    },
                    
                    genMapHTMLScriptTop = function() {  
                      
                      writeLines(paste0('  <div id="my-map"></div>'), self$outputFileHtmlCon)
                      writeLines(paste0('    <script>'), self$outputFileHtmlCon)
                      writeLines(paste0('    // This file was generated by this R program:  https://github.com/jeffblackadar/oc-transpo-maps/blob/master/oc_4_gen_html.r'), self$outputFileHtmlCon)
                      writeLines(paste0('  var map;'), self$outputFileHtmlCon)
                    },
                    
                    genMapHTMLMapTop = function() {
                      writeLines("", self$outputFileHtmlCon)
                      writeLines("      map = L.map('my-map');", self$outputFileHtmlCon)
                      writeLines("      //basemap.fitBounds(geojson.getBounds());", self$outputFileHtmlCon)
                      writeLines("      map.setView([45.42, -75.69], 10);", self$outputFileHtmlCon)
                      writeLines("      basemap.addTo(map);", self$outputFileHtmlCon)
                      writeLines("", self$outputFileHtmlCon)      
                      writeLines("      // add legend control layers - global variable with (null, null) allows indiv basemaps and overlays to be added inside functions below", self$outputFileHtmlCon)
                      writeLines("      var controlLayers = L.control.layers(null, null, {", self$outputFileHtmlCon)
                      writeLines('        position: "topleft",', self$outputFileHtmlCon)
                      writeLines("        collapsed: false // false = open by default", self$outputFileHtmlCon)
                      writeLines("      }).addTo(map);", self$outputFileHtmlCon)
                      writeLines("", self$outputFileHtmlCon)   
                    },
                    genHTMLBaseMap = function(){
                      writeLines(paste0('  window.onload = function() {'), self$outputFileHtmlCon)
                      writeLines(paste0("    var basemap = L.tileLayer('http://{s}.tile.osm.org/{z}/{x}/{y}.png', {"), self$outputFileHtmlCon)
                      writeLines(paste0("      attribution: '&copy; ",'<a href="http://osm.org/copyright">OpenStreetMap</a>',"'"), self$outputFileHtmlCon)
                      writeLines(paste0("    });"), self$outputFileHtmlCon)
                      basemapTitle="Open Street Map Today"
                    },
                    genMarker = function(marker_lat, marker_long, marker_text){
                      self$markers <- self$markers +1
                      writeLines(paste0('  var marker',paste0(self$markers),'= L.marker([',paste0(marker_lat),',',paste0(marker_long),']).addTo(map);'), self$outputFileHtmlCon)
                      writeLines(paste0('     marker',paste0(self$markers),".bindPopup('",paste0(marker_text),"').openPopup();"), self$outputFileHtmlCon)
                    },
                    
                    genHTMLEndMap = function(){
                      writeLines(paste0("      controlLayers.addOverlay(basemap, 'basemap title');"), self$outputFileHtmlCon)
                      writeLines(paste0("      "), self$outputFileHtmlCon)
                      writeLines(paste0("   // });"), self$outputFileHtmlCon)
                      writeLines(paste0("  };"), self$outputFileHtmlCon)
                      writeLines(paste0("</script>"), self$outputFileHtmlCon)
                    },
                    genHTMLEndBody = function(){
                      writeLines(paste0("<hr>"), self$outputFileHtmlCon)
                      writeLines(paste0("<p>Page last modifed: ", format(Sys.Date(),"%d %B, %Y"),"</p>"), self$outputFileHtmlCon)
                      writeLines(paste0("</body>"), self$outputFileHtmlCon)
                      writeLines(paste0("</html>"), self$outputFileHtmlCon)
                    },
                    genHTMLRailOnly = function(railYear){
                      #print(paste0("To map for ", railYear, ", adding: "))
                      #print(output_dbRows_route_style[i_route_style, 1])
                      routeLayerName <- paste0("RailOnly",railYear, "Layer")
                      
                      writeLines(paste0(""), self$outputFileHtmlCon)
                      writeLines(paste0("// layer for ",routeLayerName), self$outputFileHtmlCon)
                      writeLines(paste0("var ",routeLayerName,";"), self$outputFileHtmlCon)
                      
                      writeLines(paste0("      $.ajax({"), self$outputFileHtmlCon)
                      writeLines(paste0("        type: 'POST',"), self$outputFileHtmlCon)
                      #There are 2 files for 1954 - use the December one
                      if(railYear==1954){
                        writeLines(paste0("        url: '",railYear,"_December_rail_only_map.geojson',"), self$outputFileHtmlCon)
                      }else{
                        writeLines(paste0("        url: '",railYear,"_rail_only_map.geojson',"), self$outputFileHtmlCon)
                      }
                      writeLines(paste0("        dataType: 'json',"), self$outputFileHtmlCon)
                      writeLines(paste0("        success: function(response) {"), self$outputFileHtmlCon)
                      writeLines(paste0("          ",routeLayerName," = L.geoJson(response, {"), self$outputFileHtmlCon)
                      
                      writeLines(paste0("             onEachFeature: function (feature, layer) {"), self$outputFileHtmlCon)
                      if(railYear<=1948){
                        writeLines(paste0("                 layer.bindPopup('<h3>Route Name: '+feature.properties.RTE_NAME+'</h3><p>'+feature.properties.RTE_TYPE+'</p><p>'+feature.properties.MODE+'</p>');"), self$outputFileHtmlCon)  
                      }else{
                        writeLines(paste0("                 layer.bindPopup('<h3>Route Number: '+feature.properties.RTE_NUM+'</h3><p>'+feature.properties.RTE_TYPE+'</p><p>'+feature.properties.MODE+'</p>');"), self$outputFileHtmlCon)  
                      }
                      
                      writeLines(paste0("               }"), self$outputFileHtmlCon)
                      writeLines(paste0("             });"), self$outputFileHtmlCon)
                      writeLines(paste0("          ",routeLayerName,".setStyle({"), self$outputFileHtmlCon)
                      writeLines(paste0("            color: 'purple',"), self$outputFileHtmlCon)
                      writeLines(paste0("            weight: 2,"), self$outputFileHtmlCon)
                      writeLines(paste0('            opacity: 1'), self$outputFileHtmlCon)
                      writeLines(paste0("          });"), self$outputFileHtmlCon)
                      writeLines(paste0("          ",routeLayerName,".addTo(map);"), self$outputFileHtmlCon)
                      #Street cars ended in 1959
                      if(railYear<=1959){
                        writeLines(paste0("          controlLayers.addOverlay(",routeLayerName,", '<span style=",'"color: purple"',">",railYear," Street Cars only</span>');"), self$outputFileHtmlCon)
                      }else{
                        writeLines(paste0("          controlLayers.addOverlay(",routeLayerName,", '<span style=",'"color: purple"',">",railYear," Train only</span>');"), self$outputFileHtmlCon)
                      }
                      writeLines(paste0("        }"), self$outputFileHtmlCon)
                      writeLines(paste0("      });"), self$outputFileHtmlCon)
                      writeLines(paste0(""), self$outputFileHtmlCon)
                      
                    },
                    genHTMLUrbanExtent = function(){
                      writeLines(paste0("// Urban extent"), self$outputFileHtmlCon)
                      writeLines(paste0("      $.ajax({"), self$outputFileHtmlCon)
                      writeLines(paste0("        type: 'POST',"), self$outputFileHtmlCon)
                      writeLines(paste0("        url: 'urban_growth_future.geojson',"), self$outputFileHtmlCon)
                      writeLines(paste0("        dataType: 'json',"), self$outputFileHtmlCon)
                      writeLines(paste0("        success: function(response) {"), self$outputFileHtmlCon)
                      writeLines(paste0("          urbanExtentLayer = L.geoJson(response);"), self$outputFileHtmlCon)
                      writeLines(paste0("          urbanExtentLayer.setStyle({"), self$outputFileHtmlCon)
                      writeLines(paste0("            color: 'white',"), self$outputFileHtmlCon)
                      writeLines(paste0("            weight: 0"), self$outputFileHtmlCon)
                      writeLines(paste0("          });"), self$outputFileHtmlCon)
                      writeLines(paste0("          urbanExtentLayer.addTo(map);"), self$outputFileHtmlCon)
                      writeLines(paste0("          map.fitBounds(urbanExtentLayer.getBounds());"), self$outputFileHtmlCon)
                      writeLines(paste0("        }"), self$outputFileHtmlCon)
                      writeLines(paste0("      });"), self$outputFileHtmlCon)
                      writeLines(paste0(""), self$outputFileHtmlCon)
                    }

                  )
)
#setwd("C:\\a_orgs\\carleton\\hist3814\\R\\oc-transpo-maps\\")
outputFileHtml <- paste0("a_test_map.html")
outputFileHtmlCon<-file(outputFileHtml, open = "w")
map <- RLeafletMap$new(outputFileHtmlCon=outputFileHtmlCon)
map$genMapHTMLTop()
map$genMapHTMLScriptTop()
map$genHTMLBaseMap()
map$genMapHTMLMapTop()
map$genMarker(45.4376105533903, -75.6575202941895, '<p>The Eastview Bus Company operated in Eastview until December 29, 1950. <a href="#footnote11"><sup>11</sup></a></p><p>The route map for this bus line is not known for this year.</p>')

# map$genHTMLRailOnly(1929)
# map$genHTMLUrbanExtent()
map$genHTMLEndMap()
map$genHTMLEndBody()

close(outputFileHtmlCon)
```
